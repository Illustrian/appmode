```python
#DIN

from bs4 import BeautifulSoup 
import requests
import html5lib
import re
from urllib.parse import urlparse
import pandas as pd

url = 'https://www.zaubacorp.com/company-directors/OIL-INDIA-LIMITED/L11101AS1959GOI001148'


def cleanhtml(raw_html):
  cleanr = re.compile('<.*?>')
  cleantext = re.sub(cleanr, '', raw_html)
  return cleantext

def insert_row(idx, df, df_insert):
    return df.iloc[:idx, ].append(df_insert).append(df.iloc[idx:, ]).reset_index(drop = True)

parse_object = urlparse(url)
parse_object.netloc

r = requests.get(url) 
data = r.text
soup = BeautifulSoup(data)


# #soup.find_all('h4')
# #soup.find_all('h1')
# a=soup.find_all('h5')


# def list_slice(S, step):
#     return [S[i::step] for i in range(step)]


# a1 = list_slice(a,3)[1]    #DIN found
# #k = list_slice(k,3)
# b=a1[:-1]

# dir_url=[]
# dir_name=[]
# dir_din=[]


# for d in b:
#     dir_url.append(str(d)[13:-5].split('"> ')[0])
#     dir_name.append(cleanhtml(str(d)[13:-5].split('"> ')[1]))
    

val = []
    

table = soup.find("table", { "class" : "table table-striped" })

table_body = table.find('tbody')
rows = table_body.find_all('tr')

for row in rows:
    cells = row.findAll("td")
    column=[]
    for ele in cells:
        column.extend(abc.text for abc in ele.findAll("h5"))
        try:
            column.extend(abc.find('a').get('href') for abc in ele.findAll("h5"))
        except:
            next
        column.extend(abc.text for abc in ele.findAll("p"))

    
    if column:
        val.append(column)


df=pd.DataFrame(val,columns=['DIN','Name','Link','Position','Date'])

print (df)
```
